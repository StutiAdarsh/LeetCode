```python
from threading import Lock
class MyCircularQueue(object):
    def __init__(self, k):
        """
        :type k: int
        """       
        self.queue = [0]*k
        self.head = 0
        self.capacity = k
        self.count = 0
        self.queueLock = Lock()

    def enQueue(self, value):
        """
        :type value: int
        :rtype: bool
        """
        
        with self.queueLock:
            if self.count == self.capacity:
                return False
            self.queue[(self.head + self.count) % self.capacity] = value
            self.count += 1
        return True

    def deQueue(self):
        """
        :rtype: bool
        """
        
        if self.count == 0:
            return False
        self.head = (self.head + 1) % self.capacity
        self.count -= 1
        return True

    def Front(self):
        """
        :rtype: int
        """
        
        if self.count == 0:
            return -1
        return self.queue[(self.head)]
        

    def Rear(self):
        """
        :rtype: int
        """
        
        if self.count == 0:
            return -1
        return self.queue[(self.head + self.count - 1) % self.capacity]

    def isEmpty(self):
        """
        :rtype: bool
        """
        return self.count == 0
    
    def isFull(self):
        """
        :rtype: bool
        """
        
        return self.count == self.capacity
        


# Your MyCircularQueue object will be instantiated and called as such:
# obj = MyCircularQueue(k)
# param_1 = obj.enQueue(value)
# param_2 = obj.deQueue()
# param_3 = obj.Front()
# param_4 = obj.Rear()
# param_5 = obj.isEmpty()
# param_6 = obj.isFull()



#Time: O(1) - All of the methods in our circular data structure is of constant time complexity.
#Space: O(N) -  where N is the pre-assigned capacity of the queue
```

```python
# Create a circular queue with capacity 3
obj = MyCircularQueue(3)

# Enqueue elements
obj.enQueue(1)
obj.enQueue(2)
obj.enQueue(3)

# Dequeue an element
obj.deQueue()

# Get the front and rear elements
front = obj.Front()  # Should return 2
rear = obj.Rear()    # Should return 3
print(front, rear)

# Check if the circular queue is empty or full
isEmpty = obj.isEmpty()  # Should return False
isFull = obj.isFull()    # Should return False
print(isEmpty, isFull)
```
